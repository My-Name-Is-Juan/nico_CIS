```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(2,10,15)
Y=np.power(x,2)
Y1=Y+(np.random.random([15,])*20)
Y2=Y+(np.random.random([15,])*20)
Y3=Y+(np.random.random([15,])*20)
# print(Y1)
# print(x)
plt.plot(x,Y)
plt.plot(x,Y1)
plt.plot(x,Y2)
plt.plot(x,Y3)
plt.grid()
plt.title('line plot with data points')
plt.ylabel('variable y')
plt.xlabel('variable x')
plt.show()
```
